```python
def p2():
    def even(x):
        return x % 2 == 0
    
    # fibonacci generator
    def fib(m):
        a, b = 1, 2
        while a < m:
            yield a
            a, b = b, a + b
    
    return sum(filter(even, fib(4000000)))

p2()
```

```python
def p4():
    # number reads the same way both ways
    def palindrome(x):
        s = str(x)
        return s == s[::-1]
    
    # products of two 3 digit numbers
    def gen():
        for x in range(111, 999):
            for y in range(x, 999):
                yield x * y
            
    return max(filter(palindrome, gen()))

p4()
```

```python
import functools

def p5():
    # Euclidean greatest common divisor
    def gcd(a, b):
        if b == 0:
            return a
        else:
            return gcd(b, a % b)

    # least common multiple
    def lcm(a, b):
        return a // gcd(a, b) * b

    return functools.reduce(lcm, range(2, 20))

p5()
```
